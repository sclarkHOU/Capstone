```python
#imports
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
#read in files
aisle = pd.read_csv('data/aisles.csv')
department = pd.read_csv('data/departments.csv')
orders = pd.read_csv('data/orders.csv')
products = pd.read_csv('data/products.csv')
order_products_prior = pd.read_csv('data/order_products__prior.csv')
```

** **
## Data Exploration

```python
dataframes = ['aisle', 'department', 'orders', 'order_products_prior', 'products']

for df_name in dataframes:
    df = globals()[df_name]
    print(f"DataFrame: {df_name}")
    print("Number of rows:", len(df))
    display(df.head())
    print("Info:")
    display(df.info())
    print("Describe:")
    display(df.describe())
    print("=" * 40)
```

```python
#merging files together
df = pd.merge(order_products_prior, products, on='product_id', how='left')
df = pd.merge(df, aisle, on='aisle_id', how='left')
df = pd.merge(df, department, on='department_id', how='left')
df = pd.merge(df, orders, on='order_id', how='left')
```

```python
df.head()
```

** **
## Cleaning & Adding Columns

```python
#cleaning the data
df['product_name'] = df['product_name'].str.lower()
```

```python
#creating an organic product column 
df['organic'] = df['product_name'].str.contains('organic').astype(int)
```

```python
#creating an organic produce column
df['organic_produce'] = np.where((df['organic'] == 1) & (df['department'] == 'produce'), 1, 0)
```

```python
#creating an organic dataframe
organic = df.loc[df['organic'] ==1]
```

```python
import calendar

days=[]

for i in organic['order_dow']:
    days.append(calendar.day_name[i])
```

```python
#Adding day_of_week column 
organic['day_of_week']=days

#Adding is_weekend column
organic['is_weekend'] = organic['order_dow'].map(lambda d: 1 if d >= 5 else 0)

#Adding a day_period column
organic['day_period'] = pd.cut(organic['order_hour_of_day'], 
                                  bins=[-1, 11, 16, 23], 
                                  labels=['Morning', 'Afternoon', 'Night'], 
                                  include_lowest=True)
```

```python
organic.shape
```

## Visualizations

```python
user_cnt = organic.groupby("user_id")["order_number"].aggregate(np.max).reset_index()
user_cnt = user_cnt.order_number.value_counts()

plt.figure(figsize=(12,8))
sns.barplot(user_cnt.index, user_cnt.values, alpha=0.8, color='green')
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('Maximum order number', fontsize=12)
plt.xticks(rotation='vertical')
plt.show()
```

** **
#### Now let us see how the ordering habit changes with day of week.
** **

```python
plt.figure(figsize=(12,8))
sns.countplot(x="order_dow", data=organic, color='green')
plt.ylabel('Count', fontsize=12)
plt.xlabel('Day of week', fontsize=12)
plt.xticks(rotation='vertical')
plt.title("Frequency of order by week day", fontsize=15)
plt.show()
```

** **
#### Seems like 0 and 1 is Saturday and Sunday when the orders are high and low during Wednesday.

#### Now we i'll see what the distribution is with by time of the day.
** **

```python
plt.figure(figsize=(12,8))
sns.countplot(x="order_hour_of_day", data=organic, color='green')
plt.ylabel('Count', fontsize=12)
plt.xlabel('Hour of day', fontsize=12)
plt.xticks(rotation='vertical')
plt.title("Frequency of order by hour of day", fontsize=15)
plt.show()
```

** **
#### We see most orders are made during day. Next i'll combine the day of week and hour of day to see the distribution.
** **

```python
grouped_df = organic.groupby(["order_dow", "order_hour_of_day"])["order_number"].aggregate("count").reset_index()
grouped_df = grouped_df.pivot('order_dow', 'order_hour_of_day', 'order_number')

plt.figure(figsize=(20,10))
sns.heatmap(grouped_df)
plt.title("Frequency of Day of week Vs Hour of day")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.countplot(x="days_since_prior_order", data=organic, color='green')
plt.ylabel('Count', fontsize=12)
plt.xlabel('Days since prior order', fontsize=12)
plt.xticks(rotation='vertical')
plt.title("Frequency by days since prior order", fontsize=15)
plt.show()
```

** **
#### Customers order once in every week (check the peak at 7 days). there is a pop at day 30 but as previously mentioned, days are capped at 30 so we dont know forsure how many days are grouped in

```python
grouped_df = organic.groupby(["department_id", "aisle"])["reordered"].aggregate("mean").reset_index()

fig, ax = plt.subplots(figsize=(12,20))
ax.scatter(grouped_df.reordered.values, grouped_df.department_id.values)
for d, text in enumerate(grouped_df.aisle.values):
    ax.annotate(text, (grouped_df.reordered.values[d], grouped_df.department_id.values[d]), rotation=45, ha='center', va='center', color='green')
plt.xlabel('Reorder Ratio')
plt.ylabel('department_id')
plt.title("Reorder ratio of different aisles", fontsize=15)
plt.show()
```

** **
Next i'll look at the relationship between order of adding the product to the cart affects the reorder ratio.
** **

```python
organic["add_to_cart_order_c"] = organic["add_to_cart_order"].copy()
organic.loc[organic["add_to_cart_order_c"] > 70, "add_to_cart_order"] = 70
grouped_df = organic.groupby(["add_to_cart_order_c"])["reordered"].mean().reset_index()

plt.figure(figsize=(12,8))
sns.pointplot(grouped_df['add_to_cart_order_c'].values, grouped_df['reordered'].values, alpha=0.8, color='green')
plt.ylabel('Reorder ratio', fontsize=12)
plt.xlabel('Add to cart order', fontsize=12)
plt.title("Add to cart order - Reorder ratio", fontsize=15)
plt.xticks(rotation='vertical')
plt.show()
```

** **

#### From the chart, the products that are added to the cart at first are more likely to be reordered again compared to the ones added later.

** **

```python
grouped_df = organic.groupby(["day_of_week"])["reordered"].mean().reset_index()
grouped_df = grouped_df.sort_values(by="reordered", ascending=False)

plt.figure(figsize=(12, 8))
sns.barplot(data=grouped_df, x='day_of_week', y='reordered', alpha=0.8, color='green')
plt.ylabel('Reorder ratio', fontsize=12)
plt.xlabel('Day of week', fontsize=12)
plt.title("Reorder ratio across day of week", fontsize=15)
plt.xticks(rotation='vertical')
plt.ylim(0.5, 0.7)
plt.show()
```

```python
grouped_df = organic.groupby(["order_hour_of_day"])["reordered"].aggregate("mean").reset_index()

plt.figure(figsize=(12,8))
sns.barplot(grouped_df['order_hour_of_day'].values, grouped_df['reordered'].values, alpha=0.8, color='green')
plt.ylabel('Reorder ratio', fontsize=12)
plt.xlabel('Hour of day', fontsize=12)
plt.title("Reorder ratio across hour of day", fontsize=15)
plt.xticks(rotation='vertical')
plt.ylim(0.5, 0.7)
plt.show()
```

### Final Cleaning

```python
rows_with_nan = organic[organic['days_since_prior_order'].isna()]

sample_rows_with_nan = rows_with_nan.sample(n=5)  # Change n to the number of rows you want to display
sample_rows_with_nan
```

```python
#filling the missing values in days_since_prior_order with the median 
organic['days_since_prior_order'].fillna(organic.groupby('user_id')['days_since_prior_order'].transform(lambda x: np.round(x.median())), inplace=True)
```

```python
#with 4000 out of 10 million rows missing I will drop those
organic.dropna(subset=['days_since_prior_order'], inplace=True)
```
